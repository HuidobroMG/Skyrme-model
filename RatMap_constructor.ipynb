```python
# We use the sympy Python library to work with symbolic expressions
import sympy as sp
# Transform is needed to round the coefficients to 2 decimal places
from sympy.core.rules import Transform
# Sympy's latex printing
sp.init_printing(use_latex = True)
```

## A Rational Map in the Skyrme model maps the stereographic coordinates $z(\theta, \phi), \bar{z}(\theta, \phi)$ to the Skyrme fields

$U = \cos(f(r)) I + \sin(f(r))\vec{n}_R \vec{\tau}$,

$\vec{n}_R = \frac{1}{1+|R(z)|^2}(2\Re(R(z)), 2\Im(R(z)), 1-|R(z)|^2)$,

$R(z) = \frac{p(z)}{q(z)}$

```python
# Real and imaginary parts of the stereographic coordinate z
zRe = sp.Symbol('Re_z', real = True)
zIm = sp.Symbol('Im_z', real = True)

z = zRe + sp.I*zIm

z
```

```python
z.conjugate()
```

```python
sp.re(z), sp.im(z)
```

```python
# Rounding function up to 2 decimal places
def RoundExpr(expr):
    new_expr = expr.xreplace(Transform(lambda x: x.round(2), lambda x: isinstance(x, sp.Float)))
    return new_expr
```

```python
# Rational Map function. Inputs: Definition of R(z) = p(z)/q(z). Outputs: 
def RationalMap(Rmap_Num, Rmap_Den):
    Denom = sp.simplify(sp.expand(sp.expand(Rmap_Den)*sp.expand(Rmap_Den.conjugate())))
    Numer = sp.simplify(sp.expand(sp.expand(Rmap_Num)*sp.expand(Rmap_Den.conjugate())))
    R_Re = sp.re(Numer)
    R_Im = sp.im(Numer)
    return Denom, R_Re, R_Im
```

## B = 4 Rational Map

```python
# Define the rational map ansatz
Rmap_Num = z**4 + 2*sp.sqrt(3)*sp.I*z**2 + 1
Rmap_Den = z**4 - 2*sp.sqrt(3)*sp.I*z**2 + 1

Rmap_Num
```

```python
Denom, R_Re, R_Im = RationalMap(Rmap_Num, Rmap_Den)

Denom
```

## B = 12 Rational Map

```python
# Define the rational map ansatz
p_mas = z**4 + 2*sp.sqrt(3)*sp.I*z**2 + 1
p_menos = z**4 - 2*sp.sqrt(3)*sp.I*z**2 + 1

Rmap_Num = -0.53*p_mas**3 + 0.78*p_menos**3
Rmap_Den = p_mas**2*p_menos

Rmap_Num
```

```python
Denom, R_Re, R_Im = RationalMap(Rmap_Num, Rmap_Den)

RoundExpr(Denom)
```

```python
RoundExpr(R_Re)
```

```python
RoundExpr(R_Im)
```
